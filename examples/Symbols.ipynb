```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from metpy.plots import StationPlot
from metpy.plots.wx_symbols import current_weather, high_clouds, sky_cover
```

```python
def plot_symbols(mapper, name):
    fig = plt.figure(figsize=(2, 0.1),  dpi=300)
    ax = fig.add_axes([0, 0, 1, 1])
    ax.set_title(name, size=6)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    ax.set_frame_on(False)
    fig.patch.set_facecolor('g')
    x = np.linspace(0, 1, len(mapper))
    y = np.ones_like(x)
    codes = np.arange(len(x))
    sp = StationPlot(ax, x, y)
    sp.plot_symbol('C', codes, mapper)
    sp.plot_parameter('S', codes, size=4)
    ax.set_xlim(0, 1)
    ax.set_ylim(0.5, 1.3)
```

```python
plot_symbols(sky_cover, "Sky Cover Symbols")
```

```python
plot_symbols(high_clouds, "High Cloud Symbols")
```

```python
# Get the mapper that converts a code for current weather (from WMO) to the appropriate
# unicode code point
from metpy.plots.wx_symbols import current_weather, wx_symbol_font

# Need to make a new copy of the font so we can make it bigger
big_font = wx_symbol_font.copy()
big_font.set_size(36)

# Create a plot to loop over all of the codes and display it
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
for i in range(100):
    ax.text(i % 10, i // 10, current_weather(i), fontproperties=big_font)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10);
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 2))
for i in range(10):
    ax.text(i, 0, sky_cover(i), fontproperties=big_font)
ax.set_xlim(0, 10)
ax.set_ylim(0, 1);
```
